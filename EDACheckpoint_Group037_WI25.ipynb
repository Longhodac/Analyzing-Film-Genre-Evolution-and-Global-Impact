```python
# For Pre Pandemic
# Ensure genres are lists by splitting them at ", "
pre_pandemic_movies["genres"] = pre_pandemic_movies["genres"].str.split(", ")

# Use explode to create separate rows for each genre
pre_pandemic_movies_exploded = pre_pandemic_movies.explode("genres", ignore_index=True)
```

```python
# Count occurrences of each genre
genre_counts_pre = pre_pandemic_movies_exploded["genres"].value_counts().reset_index()

# Rename columns for clarity
genre_counts_pre.columns = ["Genre", "Occurrences"]
```

```python
# For Post Pandemic
# Ensure genres are lists by splitting them at ", "
post_pandemic_movies["genres"] = post_pandemic_movies["genres"].str.split(", ")

# Use explode to create separate rows for each genre
post_pandemic_movies_exploded = post_pandemic_movies.explode("genres", ignore_index=True)
```

```python
# Count occurrences of each genre
genre_counts_post = post_pandemic_movies_exploded["genres"].value_counts().reset_index()

# Rename columns for clarity
genre_counts_post.columns = ["Genre", "Occurrences"]
```

```python
import statsmodels.api as sm

# Add a column to indicate period (0 for pre-pandemic, 1 for post-pandemic)
genre_counts_pre["Period"] = 0  # Pre-pandemic indicator
genre_counts_post["Period"] = 1  # Post-pandemic indicator

# Combine both datasets
df_genre_trends = pd.concat([genre_counts_pre, genre_counts_post])

# Independent variable (Period) and dependent variable (Occurrences)
X = df_genre_trends["Period"]
y = df_genre_trends["Occurrences"]

# Add a constant for the intercept
X = sm.add_constant(X)

# Fit the linear model
model = sm.OLS(y, X).fit()

# Display summary of the regression
print(model.summary())
```

## Results from this:
 ### 1. Coefficient for Period: -1.4853
- Suggests that, on average, genres appear slightly less frequently in the top 100 highest-grossing films post- pandemic compared to pre-pandemic. 
- The effect size is small (about -1.49 occurrences on average).
- **No strong evidence** that the pandemic caused a major shift in genre frequency among top-grossing films.

### 2. P-value for Period: 0.776
- Since p = 0.776 is much higher than 0.05, we fail to reject the null hypothesis.
- Change in genre frequency could be due to *random variation* rather than a real trend.
- **No statistically significant evidence** that the pandemic changed genre popularity.

### 3. R-squared: 0.003
- Means that only 0.3% of the variation in genre frequency is explained by the pandemic period (pre vs. post).
- Extremely low value, meaning that *other factors beyond the pandemic must affecting changes in genre trends*.
- So, the pandemic itself **does not appear** to be a strong predictor of changes in genre frequency.

### 4. Confidence Interval: [-12.021, 9.050]
- Since it includes both negative and positive values (-12.021 to 9.050), the effect could be strongly negative, positive, or nonexistent.
- This wide range means we *lack precise evidence* about how the pandemic impacted genre frequency.
- Because the interval crosses zero, **we cannot conclude that the pandemic definitely impacted genres** in a meaningful way.

## Final Conclusions
This analysis suggests that the COVID-19 pandemic ***did not significantly impact the distribution of genres in the top 100 highest-grossing films***.

## Extract Top 15 Words
First, we will extract the top 15 words from both pre- and post-pandemic movies DataFrames for comparison. These words will come from the 'Overview' and 'Keywords' columns found in the original dataset.

```python
# Function to extract top 15 words

def top_words(text_series):
    vectorizer = CountVectorizer(stop_words='english')
    matrix = vectorizer.fit_transform(text_series.dropna())
    
    word_counts = matrix.sum(axis=0)
    words_freq = [(word, word_counts[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse = True)
    
    return words_freq[:15]
```

### Sentiment Analysis
Through utilizing sentiment analysis, we aim to examine the sentiment of movie descriptions before and after the COVID-19 pandemic. We want to assess whether the tone of movie overviews and keywords have shifted before and after the pandemic, potentially reflecting changes in the kinds of movies that do well based on revenue. We want to categorize these words based on whether they are negative, neutral, or positive, comparing sentiment distributions across movies from both time periods. This approach provides insights on if the emotional tone of films has evolved.

```python
# Function to calculate sentiment score of words
def sentiment_score(word_set):
    return {word: TextBlob(word).sentiment.polarity for word in word_set}
```

```python
#Convert lists to sets for comparison (unique values)
pre_overview_set = set([word for word, _ in pre_overview_top])
post_overview_set = set([word for word, _ in post_overview_top])

# Compute sentiment scores using function
pre_overview_sentiments = sentiment_score(pre_overview_set)
post_overview_sentiments = sentiment_score(post_overview_set)

# Convert to DataFrame
pre_overview_sent_df = pd.DataFrame(pre_overview_sentiments.items(), columns=['Word', 'Sentiment'])
post_overview_sent_df = pd.DataFrame(post_overview_sentiments.items(), columns=['Word', 'Sentiment'])
```

```python
from collections import Counter

def simpsons_diversity_index(genre_counts):
    total_count = sum(genre_counts.values())
    index = 1 - sum((count / total_count) ** 2 for count in genre_counts.values())
    return index
```

```python
#Calculates genre counts
genre_counts_pre = Counter(pre_genres_movies['genre'])

#Uses the counts to calculate Simpson's Diversity Index
diversity_index_pre = simpsons_diversity_index(genre_counts_pre)
print(f"Simpson's Diversity Index for Pre Pandemic Movies: {diversity_index_pre}")
```

```python
#Calculates genre counts
genre_counts_post = Counter(post_genres_movies['genre'])

#Uses the counts to calculate Simpson's Diversity Index
diversity_index_post = simpsons_diversity_index(genre_counts_post)
print(f"Simpson's Diversity Index for Post Pandemic Movies: {diversity_index_post}")
```

Simpson's Diversity Index (SDI) is a measure used to quantify the diversity of a dataset. In this case, it helps us understand the diversity of movie genres. The SDI value ranges between 0 and 1:
- A **higher value** (close to 1) indicates **lower diversity**, meaning most of the movies belong to a small number of genres.
- A **lower value** (close to 0) indicates **higher diversity**, meaning the genres are more evenly distributed across the dataset.

### Pre Pandemic Movies: 0.896
- An SDI of **0.896** suggests that the genre distribution for **pre-pandemic movies** is somewhat concentrated around a few dominant genres.
- This could be seen with a higher concentration of both **Adventure** and **Action** movies.

### Post Pandemic Movies: 0.909
- The SDI for **post-pandemic movies** is **0.909**, which is slightly higher than the pre-pandemic value.
- This indicates that the **post-pandemic movie genre distribution** is **slightly more concentrated**, with fewer diverse genres than the pre-pandemic period.
- In the graph, we can see that **Action** movies are the **only** dominant genre in post-pandemic movies, while pre-andemic movies have **two** dominant genres.

### Conclusion:
- While both periods show a **relatively high SDI**, indicating a concentration of movie genres, the **post-pandemic period** shows a **slightly more concentrated genre distribution** than the pre-pandemic period.
- There was a slight shift in genre diversity through the COVID-19 pandemic with **Action** remaining a dominant genre while **Adventure** taking a small decline.
- Even though the post-pandemic movie set had an **extra** genre, it wasn't enough to lower its SDI value.

Using revenue as a metric for movie success may disproportionately highlight blockbuster films while overlooking artistic and cultural contributions from non-mainstream cinema.
