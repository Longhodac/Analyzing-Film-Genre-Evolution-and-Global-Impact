## Section 1: Looking at Overall Genre Trends using a Linear Model

#### Preparing Data for Analysis

```python
# Make copies of the dataframes before dropping columns for linear model
pre_pandemic_movies_text = pre_pandemic_movies.copy()
post_pandemic_movies_text = post_pandemic_movies.copy()
```

```python
# Drop columns not needed for linear model for pre pandemic
pre_pandemic_movies.drop(columns=['overview', 'keywords'], inplace=True)
```

```python
# Drop columns not needed for linear model for post pandemic
post_pandemic_movies.drop(columns=['overview', 'keywords'], inplace=True)
```

```python
# For Pre Pandemic
# Ensure genres are lists by splitting them at ", "
pre_pandemic_movies["genres"] = pre_pandemic_movies["genres"].str.split(", ")

# Use explode to create separate rows for each genre
pre_pandemic_movies_exploded = pre_pandemic_movies.explode("genres", ignore_index=True)
```

```python
# Count occurrences of each genre
genre_counts_pre = pre_pandemic_movies_exploded["genres"].value_counts().reset_index()

# Rename columns for clarity
genre_counts_pre.columns = ["Genre", "Occurrences"]
```

```python
# For Post Pandemic
# Ensure genres are lists by splitting them at ", "
post_pandemic_movies["genres"] = post_pandemic_movies["genres"].str.split(", ")

# Use explode to create separate rows for each genre
post_pandemic_movies_exploded = post_pandemic_movies.explode("genres", ignore_index=True)
```

```python
# Count occurrences of each genre
genre_counts_post = post_pandemic_movies_exploded["genres"].value_counts().reset_index()

# Rename columns for clarity
genre_counts_post.columns = ["Genre", "Occurrences"]
```

#### Summary of the Regression

```python
# Add a column to indicate period (0 for pre-pandemic, 1 for post-pandemic)
genre_counts_pre["Period"] = 0  # Pre-pandemic indicator
genre_counts_post["Period"] = 1  # Post-pandemic indicator

# Combine both datasets
df_genre_trends = pd.concat([genre_counts_pre, genre_counts_post])

# Independent variable (Period) and dependent variable (Occurrences)
X = df_genre_trends["Period"]
y = df_genre_trends["Occurrences"]

# Add a constant for the intercept
X = sm.add_constant(X)

# Fit the linear model
model = sm.OLS(y, X).fit()

# Display summary of the regression
print(model.summary())
```

#### Regression Results

1. Coefficient for Period: -1.4853  
    - Suggests that, on average, genres appear slightly less frequently in the top 100 highest-grossing films post- pandemic compared to pre-pandemic. 
    - The effect size is small (about -1.49 occurrences on average).
    - **No strong evidence** that the pandemic caused a major shift in genre frequency among top-grossing films.

2. P-value for Period: 0.776  
    - Since p = 0.776 is much higher than 0.05, we fail to reject the null hypothesis.
    - Change in genre frequency could be due to *random variation* rather than a real trend.
    - **No statistically significant evidence** that the pandemic changed genre popularity.

3. R-squared: 0.003 
    - Means that only 0.3% of the variation in genre frequency is explained by the pandemic period (pre vs. post).
    - Extremely low value, meaning that *other factors beyond the pandemic must affecting changes in genre trends*.
    - So, the pandemic itself **does not appear** to be a strong predictor of changes in genre frequency.

4. Confidence Interval: [-12.021, 9.050]  
    - Since it includes both negative and positive values (-12.021 to 9.050), the effect could be strongly negative, positive, or nonexistent.
    - This wide range means we *lack precise evidence* about how the pandemic impacted genre frequency.
    - Because the interval crosses zero, **we cannot conclude that the pandemic definitely impacted genres** in a meaningful way.

## Section 2: Paired T-Test Analysis of Genre Distribution  
In this section, we will conduct a paired t-test to determine if there are statistically significant differences in the distribution of genres between pre-pandemic (2017-2019) and post-pandemic (2022-2024) top-grossing films. This analysis will help us evaluate whether the COVID-19 pandemic had a measurable impact on the popularity of specific film genres.

```python
# Prepare our data for the paired t-test by merging the genre counts
genre_comparison = pd.merge(genre_counts_pre, genre_counts_post, on="Genre", how="outer", suffixes=('_pre', '_post'))
```

```python
# Fill NaN values with 0 (for genres that appear in one period but not the other)
genre_comparison = genre_comparison.fillna(0)
```

```python
genre_comparison.head(10)
```

Before conducting the t-test, let's examine the descriptive statistics for genre occurrences in both time periods to get a better understanding of our data.

```python
print(genre_comparison['Occurrences_pre'].describe())
print(genre_comparison['Occurrences_post'].describe())
```

#### Visualizing Genre Differences
Let's calculate and visualize the differences in genre occurrences between the post-pandemic and pre-pandemic periods. This will help us identify which genres have seen the biggest changes.

```python
# Calculate differences for each genre
genre_comparison['Difference'] = genre_comparison['Occurrences_post'] - genre_comparison['Occurrences_pre']
```

```python
# Visualize the differences
plt.figure(figsize=(12, 6))
sns.barplot(x='Genre', y='Difference', data=genre_comparison.sort_values('Difference', ascending=False))
plt.title('Difference in Genre Occurrences (Post-Pandemic minus Pre-Pandemic)')
plt.xticks(rotation=90)
plt.axhline(y=0, color='red', linestyle='-')
plt.tight_layout()
plt.show()
```

#### Results
The graph reveals significant shifts in film genre popularity between pre-pandemic (2017-2019) and post-pandemic (2022-2024) periods. 

- Thrillers showed the most dramatic increase (+13 occurrences), followed by Mystery (+9), while Adventure experienced the steepest decline (-19), with Fantasy and Family films also decreasing substantially (-12 to -13). This suggests a clear trend away from family-oriented and escapist content toward more mature, suspenseful narratives.   
- The rise in emotionally intense genres like Crime, Drama, and Horror alongside the decline in traditionally big-budget genres (Adventure, Science Fiction) may reflect both changing audience preferences and industry adjustments following the pandemic, with studios potentially redirecting family content to streaming platforms while reserving theatrical releases for genres that benefit more from the big-screen experience.

### Paired T-Test Analysis
Now, we'll perform a paired t-test to determine if there's a statistically significant difference in genre occurrences before and after the pandemic. Our null hypothesis is that there is no difference in genre distribution between the two time periods.

```python
# Perform paired t-test
t_value, p_value = stats.ttest_rel(genre_comparison['Occurrences_pre'], genre_comparison['Occurrences_post'])

print(f"Paired t-test results:")
print(f"t-value: {t_value:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("There is a statistically significant difference in genre distribution before and after the pandemic.")
else:
    print(f"Since p-value ({p_value:.4f}) is greater than alpha ({alpha}), we fail to reject the null hypothesis.")
    print("There is no statistically significant difference in genre distribution before and after the pandemic.")
```

### T-Test Results
The paired t-test results (t-value: 0.1850, p-value: 0.8555) indicate there is no statistically significant difference in genre occurrences between pre-pandemic and post-pandemic films. 
- The high p-value (0.8555) far exceeds the conventional significance threshold of 0.05, meaning we fail to reject the null hypothesis. Despite the visible differences in specific genres shown in your previous graph, these changes appear to balance each other out when considering the overall genre landscape. This suggests that while individual genres may have shifted in popularity (with thriller and mystery increasing, adventure and family decreasing), the collective distribution of genres remained statistically similar. 
- The low t-value (0.1850) further confirms the similarity between the two time periods. 

These results indicate that the pandemic did not cause a statistically significant systematic shift in the overall genre distribution of top-grossing films, even though specific genres experienced notable changes in either direction.

## Section 3: Simpson's Diversity Index

The code used in this section was written with the assistance of GenAI.

```python
# Split genres by comma into their own rows
pre_genres = pre_pandemic_movies_text['genres'].str.split(', ', expand=True).stack().reset_index(drop=True, level=1)
pre_genres.name = 'genre'

# Create a new dataframe using the individualized genres
pre_genres_movies = pd.DataFrame({
    'title': pre_pandemic_movies_text['title'].repeat(
        pre_pandemic_movies_text['genres'].str.split(', ').str.len()
    ),
    'genre': pre_genres
})

pre_genres_movies.head()
```

```python
# Split genres by comma into their own rows
post_genres = post_pandemic_movies_text['genres'].str.split(', ', expand=True).stack().reset_index(drop=True, level=1)
post_genres.name = 'genre'

# Create a new dataframe using the individualized genres
post_genres_movies = pd.DataFrame({
    'title': post_pandemic_movies_text['title'].repeat(
        post_pandemic_movies_text['genres'].str.split(', ').str.len()
    ),
    'genre': post_genres
})

post_genres_movies.head()
```

```python
def simpsons_diversity_index(genre_counts):
    total_count = sum(genre_counts.values())
    index = 1 - sum((count / total_count) ** 2 for count in genre_counts.values())
    return index
```

```python
# Calculates genre counts
genre_counts_pre = Counter(pre_genres_movies['genre'])

#Uses the counts to calculate Simpson's Diversity Index
diversity_index_pre = simpsons_diversity_index(genre_counts_pre)
print(f"Simpson's Diversity Index for Pre Pandemic Movies: {diversity_index_pre}")
```

```python
# Calculates genre counts
genre_counts_post = Counter(post_genres_movies['genre'])

# Uses the counts to calculate Simpson's Diversity Index
diversity_index_post = simpsons_diversity_index(genre_counts_post)
print(f"Simpson's Diversity Index for Post Pandemic Movies: {diversity_index_post}")
```

#### Interpretation:
There is a slight increase in the diversity index post-pandemic, suggesting that movie genres became slightly more varied in the highest-grossing films.
The difference, though not drastic, implies a shift in audience preferences and studio productions post-pandemic, possibly due to changing viewing habits and increased experimentation with diverse genres.

#### Extract Top 15 Words
First, we will extract the top 15 words from both pre- and post-pandemic movies DataFrames for comparison. These words will come from the 'Overview' and 'Keywords' columns found in the original dataset.

```python
# Function to extract top 15 words

def top_words(text_series):
    vectorizer = CountVectorizer(stop_words='english')
    matrix = vectorizer.fit_transform(text_series.dropna())
    
    word_counts = matrix.sum(axis=0)
    words_freq = [(word, word_counts[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse = True)
    
    return words_freq[:15]
```

#### Compare Keywords Between Periods

```python
from collections import Counter

def clean_keywords(keyword_string):
    """Clean and process the keyword string into a list of individual keywords."""
    if pd.isna(keyword_string):
        return []
    
    # Split by comma and strip whitespace
    keywords = [k.strip() for k in keyword_string.split(',')]
    return keywords

# Create explicit copies of your DataFrames first
pre_top_100_plot = pre_pandemic_movies_text.copy()
post_top_100_plot = post_pandemic_movies_text.copy()

# Now add the new column using .loc
pre_top_100_plot.loc[:, 'keywords'] = pre_top_100_plot['keywords'].apply(clean_keywords)
post_top_100_plot.loc[:, 'keywords'] = post_top_100_plot['keywords'].apply(clean_keywords)

def get_keyword_counts(keyword_lists):
    """Count the frequency of each keyword across all movies."""
    all_keywords = []
    for keywords in keyword_lists:
        all_keywords.extend(keywords)
    
    return Counter(all_keywords)

# Get keyword counts for both datasets
pre_keyword_counts = get_keyword_counts(pre_top_100_plot['keywords'])
post_keyword_counts = get_keyword_counts(post_top_100_plot['keywords'])

def compare_keyword_frequencies(pre_counts, post_counts, top_n=20):
    """Compare keyword frequencies between pre and post periods."""
    # Get top keywords from both periods
    top_keywords = set()
    for keyword, _ in pre_counts.most_common(top_n):
        top_keywords.add(keyword)
    for keyword, _ in post_counts.most_common(top_n):
        top_keywords.add(keyword)
    
    # Create a comparison dataframe
    comparison = pd.DataFrame(index=list(top_keywords), columns=['Pre-2020', 'Post-2020'])
    
    for keyword in top_keywords:
        comparison.loc[keyword, 'Pre-2020'] = pre_counts.get(keyword, 0)
        comparison.loc[keyword, 'Post-2020'] = post_counts.get(keyword, 0)
    
    # Calculate percentage change
    comparison['Absolute_Change'] = comparison['Post-2020'] - comparison['Pre-2020']
    comparison['Percent_Change'] = (comparison['Absolute_Change'] / comparison['Pre-2020'] * 100).fillna(float('inf'))
    
    # Sort by absolute change
    comparison = comparison.sort_values('Absolute_Change', ascending=False)
    
    return comparison

# Generate comparison
keyword_comparison = compare_keyword_frequencies(pre_keyword_counts, post_keyword_counts)

# Display the comparison
print("\nKeyword frequency comparison (sorted by absolute change):")
print(keyword_comparison.head(10))
```

#### Visualize Keyword Changes

```python
both_periods = keyword_comparison[(keyword_comparison['Pre-2020'] > 0) & 
                                  (keyword_comparison['Post-2020'] > 0)].copy()

# Explicitly convert columns to numeric types
both_periods['Pre-2020'] = pd.to_numeric(both_periods['Pre-2020'], errors='coerce')
both_periods['Post-2020'] = pd.to_numeric(both_periods['Post-2020'], errors='coerce')

# Calculate percentage for visualization
both_periods['Pre_Percentage'] = both_periods['Pre-2020'] / both_periods['Pre-2020'].sum() * 100
both_periods['Post_Percentage'] = both_periods['Post-2020'] / both_periods['Post-2020'].sum() * 100
both_periods['Percentage_Change'] = both_periods['Post_Percentage'] - both_periods['Pre_Percentage']

# Make sure Percentage_Change is numeric
both_periods['Percentage_Change'] = pd.to_numeric(both_periods['Percentage_Change'], errors='coerce')

# Replace append() with concat() since append() is deprecated
from pandas import concat
top_increases = both_periods.nlargest(10, 'Percentage_Change')
top_decreases = both_periods.nsmallest(10, 'Percentage_Change')
top_changes = concat([top_increases, top_decreases])

# Plot with proper data
plt.figure(figsize=(12, 10))
bars = plt.barh(top_changes.index, top_changes['Percentage_Change'], 
                color=['g' if x > 0 else 'r' for x in top_changes['Percentage_Change']])
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Percentage Point Change')
plt.title('Keywords with Biggest Changes (Pre-2020 to Post-2020)')
plt.tight_layout()
plt.savefig('keyword_changes.png')
plt.show()
```

#### Analyze Keyword Co-Occurrence

```python
def generate_keyword_pairs(keyword_lists):
    """Generate all pairs of co-occurring keywords."""
    pairs = []
    for keywords in keyword_lists:
        # Generate all possible pairs in this movie
        if len(keywords) > 1:
            for i in range(len(keywords)):
                for j in range(i+1, len(keywords)):
                    # Sort keywords alphabetically for consistent counting
                    pair = tuple(sorted([keywords[i], keywords[j]]))
                    pairs.append(pair)
    
    return Counter(pairs)

# Get co-occurrence counts
pre_cooccurrence = generate_keyword_pairs(pre_top_100_plot['keywords'])
post_cooccurrence = generate_keyword_pairs(post_top_100_plot['keywords'])

# Display top co-occurring keywords
print("\nTop 10 co-occurring keywords in pre-2020 movies:")
for pair, count in pre_cooccurrence.most_common(10):
    print(f"{pair[0]} + {pair[1]}: {count}")

print("\nTop 10 co-occurring keywords in post-2020 movies:")
for pair, count in post_cooccurrence.most_common(10):
    print(f"{pair[0]} + {pair[1]}: {count}")
```

```python
# Function to calculate sentiment score of words
def sentiment_score(word_set):
    return {word: TextBlob(word).sentiment.polarity for word in word_set}
```
